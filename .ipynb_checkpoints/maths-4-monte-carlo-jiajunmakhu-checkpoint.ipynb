# Monte Carlo simulatie
In dit project werd als opdracht gegeven om een Monte Carlo simulatie te gaan maken met behulp van je eigen gemaakte pseudo random number generators. Deze PRNG's staan in de python file RNGs gedefinieerd. De casus die bij dit opdracht hoort gaat over de eredivisie. Er wordt dus gevraagd om een voorspelling te maken aan de hand van de gegeven win/gelijkspel/verlies-kansen voor het eerste deel van de opdracht en voor het tweede deel aan de hand van de gemiddelde doelsaldo. Deze voorspelling wordt meerdere (duizenden) malen gemaakt om een voorspelde percentage te krijgen dat een bepaalde partij op een bepaalde plaats staat in de ranking. Hoe meer een voorspelling wordt gedaan, hoe accurater de resultaten zullen zijn.

Er zijn twee verschillende Pseudo Random Number Generators gemaakt. Deze zijn de Linear Congruential Generator en Lagged Fibonacci Generator. Linear Congruential Generator is geimplementeerd aan de hand van de onderstaande formule:
\
$X_{n} = (a \cdot X_{n-1} + c) \mod m, 0 < a < m, 0 \le c < m$

Hierin zijn de $a$ en $c$ kleiner dan de $m$, waarbij $c$ een nul kan zijn en $a$ niet. De $X_{n-1}$ in deze formule is de vorige berekende waarde met de LCG. Voor de eerste instantie wordt de gegeven seed gebruikt. Voor de implementatie van de LGC wordt de onderstaande formule gebruikt:
\
$X_n = X_{n-j} \star X_{n-k}  \mod m,   0 < j < k$

In deze formule gaat de $j$ en $k$ over de vorige $j$e en $k$e waarde die berekend werd door de LFG. Hier moet de $j$ kleiner zijn dan de $k$. De \star is een meegegeven operator. Deze zijn optellen, aftrekken, vermenigvuldigen of de XOR.

```python
# De configuraties van Microsoft Visual wordt gepakt voor de LCG bron: https://en.wikipedia.org/wiki/Linear_congruential_generator
lcg = RNGs.LCG(12345, 214013, 2531011, 2**32)

# De configuraties werden van de wikipedie gepakt voor LFG bron: https://en.wikipedia.org/wiki/Lagged_Fibonacci_generator
# Hier werd de vermenigvuldiging gebruikt voor de LFG
lfg = RNGs.LFG(12345, 9739, 23209, 2**31 + 1, lambda x, y: x * y)
```

Om te kijken of de zelf gecodeerde pseudo random generator werkt of niet, wordt deze meerdere keren gerund en de waarden daarvan worden in een lineplot gezet. Hierdoor kan er gekeken worden of er een patroon te vinden is in de berekende waarden. Als er snel een patroon te vinden is, dan zal de rng niet heel effectief zijn.

```python
# Lijst voor de gegenereerde waarden
lcg_list = []
lfg_list = []
steps = []
for i in range(0, 100):
    lcg_list.append(lcg.next())
    lfg_list.append(lfg.next())
    steps.append(i)
```

Door te kijken naar de plot geproduceerde waarden over de iteraties, is er te zien dat er een willekeur is. In dit geval, is er geen patroon hier te vinden, wat dus betekend dat er voldoende willekeur meegenomen is voor beide lcf en lfg.

```python
plt.plot(steps, lcg_list)
plt.title("lcg values")
plt.show()
```

```python
plt.plot(steps, lfg_list)
plt.title("lfg values")
plt.show()
```

Hieronder staan de meegegeven tabellen. De eerste is de win-gelijk-verlies kans tabel voor de home vs out partijen. Elke index van deze dictionary geeft als key de partij dat thuis speelt met als waarde de dictionary voor de win-gelijk kansen tegen elke uit partij. Bijvoorbeeld percentage_tabel["PSV"]["Feyenoord"] geeft PSV vs Feyenoord, met PSV als thuis en Feyenoord als uit en geeft (0.54, 0.22) als de kansen. De tweede tabel is de gemiddelde doelpunten die een team scoort of tegen krijgt bij thuis of uit situatie. Elke index geeft als key de partij en als value de waarde van elke situatie. Hierin zijn: "HS": aantal gescoord in thuis, "HC": doelpunten tegen in thuis, "AS": aantal gescoord uit, "AC": doelpunten tegen uit. Als er gekeken wordt naar de percentages van elke team, dan is er te zien dat Ajax de hoogste percentages heeft om te winnen tegen elk andere team. De verwachting zal dan ook zijn dat Ajax bovenaan zal staan na het runnen van alle wedstrijden. Ajax zal dan ook naar verwachting een hoge percentage krijgen op de eerste plek. Willem II zal dan wel de laagste zijn als er gekeken wordt naar de percentages.

```python
percentage_table = {"Ajax": {"Feyenoord": (0.65, 0.17), "PSV": (0.54, 0.21), "FC Utrecht": (0.74, 0.14), "Willem II": (0.78, 0.13)}, 
                    "Feyenoord": {"Ajax": (0.30, 0.21), "PSV": (0.37, 0.24), "FC Utrecht": (0.51, 0.22), "Willem II": (0.60, 0.21)}, 
                    "PSV": {"Ajax": (0.39, 0.22), "Feyenoord": (0.54, 0.22), "FC Utrecht": (0.62, 0.20), "Willem II": (0.62, 0.22)}, 
                    "FC Utrecht": {"Ajax": (0.25, 0.14), "Feyenoord": (0.37, 0.23), "PSV": (0.29, 0.24), "Willem II": (0.52, 0.23)}, 
                    "Willem II": {"Ajax": (0.17, 0.18), "Feyenoord": (0.20, 0.26), "PSV": (0.23, 0.24), "FC Utrecht": (0.37, 0.25)}}
```

Als er gekeken wordt naar de gemiddelde doelpunten van elke team, is er te zien dat hier ook Ajax hoogstwaarschijnlijk bovenaan zult staan. Dit komt door de hoge gemiddelde die Ajax zal scoren, namelijk voor "HS", dus thuis, is dat 3.2 en voor "AS", dus uit, is dit 3.1. Niet alleen zullen zij veel scoren, maar ook zullen zij weinig doelpunten tegen krijgen.

```python
lambda_table = {"Ajax": {"HS": 3.2, "HC": 0.9, "AS": 3.1, "AC": 0.6},
                "Feyenoord": {"HS": 2.4, "HC": 1.1, "AS": 2.2, "AC": 0.8},
                "PSV": {"HS": 2.1, "HC": 0.7, "AS": 1.8, "AC": 1.3},
                "FC Utrecht": {"HS": 1.9, "HC": 1.2, "AS": 3.0, "AC": 2.4},
                "Willem II": {"HS": 1.4, "HC": 1.7, "AS": 1.0, "AC": 1.5}}
```

```python
# De gebruikte RNG
rng = lfg
```

Onderstaand staat een functie dat kijkt welke team gewonnen heeft gebaseerd op de gegeven input. Deze input is de waarde gegenereerd door de RNG, wat tussen de 0 en 1 zit. Deze functie is gebaseerd op de cumulatieve winst en gelijkspel kansen van de wedstrijden. Dit betekent dat de kansen plus elkaar wordt gedaan oplopend tot 1, wat 100 procent voorstelt. Deze wordt hier geintegreerd zo dat als de input tussen 0 en de winkans zit, dat degene van thuis gewonnen heeft en dus 3 punten krijgt. Voor gelijkspel is dit een waarde tussen de winkans en de winkans plus gelijkkans. Als het ook daar buiten valt, maar binnen 1 valt, dan heeft de team thuis verloren en krijgt dus 0 punten.

```python
def win_status_check(input: float, win: float, draw: float) -> tuple[int, int]:
    """
    Checkt of een team gewonnen heeft of niet gebaseerd op de gegeven input.
    :param input: The random nummer waarop de win status is gebaseerd
    :param win: Percentage gewonnen
    :param draw: Percentage gelijkspel
    :return:
    Returnt 3 voor een win, 1 voor een gelijkspel en 0 voor verloren. In een tuple
    gereturned waarin de eerste index voor de thuispartij geldt en de tweede voor de uit partij
    """
    if input < win:
        return 3, 0
    elif input < win + draw:
        return 1, 1
    # If the team didn't win or draw, then the team lost
    else:
        return 0, 3
```

Hieronder staat een functie die de simulatie een keer runt. Dit houdt in dat er gesimuleerd wordt dat elke team tegen alle andere teams gespeeld heeft. Hierin zijn de resultaten van elke wedstrijd gebaseerd op de eerder gemaakte RNG en de win-gelijk kansen. Deze geeft de dictionary terug van de punten voor elke team.

```python
def run_simulation() -> dict:
    """
    Deze functie runt de simulatie van een reeks aan wedstrijden een keer. De winstanden worden hier met een kans op win, gelijk of verlies kans bepaald.
    :return:
    De dictionary waarin de key de naam van de team is en het aantal punten als value geeft.
    """
    points = {"Ajax": 0, "Feyenoord": 0, "PSV": 0, "FC Utrecht": 0, "Willem II": 0}
    for home, out_chances in percentage_table.items():
        for out, chances in out_chances.items():
            result_home, result_out = win_status_check(rng.next(), chances[0], chances[1])
            points[home] += result_home
            points[out] += result_out
    return points
```

De onderstaande functie geeft de positie voor elke team weer. Deze wordt meerdere keren gerund om daaruit een percentage te krijgen van hoevaak een bepaalde team op welke plaats staat. Deze zijn ook gelijk de kansen op een bepaalde plek op de rangschikking voor elke team.

```python
def positions(points:dict) -> dict:
    """
    Deze functie sorteert de teams gebaseerd op het behaalde punten in aflopende volgorde.
    Vervolgens geeft deze functie elke team hun positie op de rangschikking.
    :param points: De dictionary met de punten van elke team
    :return:
    De dictionary waarin de positie van elke team staat.
    """
    # Sorteert de teams gebaseerd op het aantal punten dat ze hebben gehaald.
    sorted_points = sorted(points.items(), key=lambda x:x[1], reverse=True)
    positions_teams = {}
    for i in range(0,len(sorted_points)):
        positions_teams[sorted_points[i][0]] = i + 1
    return positions_teams
```

```python
def ranking_chances(runs: int) -> pd.DataFrame:
    """
    Deze functie runt de simulatie een aantal keer gelijk aan de runs. Voor elke run worden de posities bijgehouden van elke team na de reeks aan wedstrijden. Na het aantal runs worden de percentages berekend voor hoevaak een team op een bepaalde plek heeft gestaan.
    :param runs: Aantal runs van een simulatie
    :return:
    De dataframe waarin de percentages staan van het aantal keren dat een team op een bepaalde plek heeft gestaan in de afgelopen runs. Deze staat in aflopende volgorde gebaseerd op de eerste plek kans.
    """
    past_positions = {"Ajax": [], "Feyenoord": [], "PSV": [], "FC Utrecht": [], "Willem II": []}
    df = pd.DataFrame(columns=["Team", "1st", "2nd", "3rd", "4th", "5th"])
    # Runt de simulatie een aantal keren gelijk aan het gegeven parameter
    for _ in range(0, runs):
        points = run_simulation()
        team_positions = positions(points)
        for team, position in team_positions.items():
            past_positions[team].append(position)
    # Berekent de percentages voor elke team voor elke positie
    for team, past_pos in past_positions.items():
        df.loc[len(df)] = [team, round(past_pos.count(1)/runs * 100, 1), round(past_pos.count(2)/runs * 100, 1), round(past_pos.count(3)/runs * 100, 1), round(past_pos.count(4)/runs * 100, 1), round(past_pos.count(5)/runs * 100, 1)]
    return df.sort_values(by='1st', ascending=False)
```

De simulatie wordt hier tien duizend keer gerund. Hieronder staan de kansen voor elke team. Hier is te zien dat Ajax de grootste kans heeft om te winnen. Dit is volgens verwachting als er alleen gekeken werd naar de percentages in de tabel. Ook dat Willem II het laagste uit zal komen is hier terug te zien.

```python
ranking_chances(10000)
```

Voor het tweede deel van de opdracht moeten de doelsaldo gebruikt worden om te kijken wie er gewonnen heeft in de plaats van een willekeurig kans. Dit wordt aan de hand van de poisson verdeling gedaan. Hiermee worden de kansen berekend op een bepaald aantal doelpunten. De poisson formule is hier direct onder gedefinieerd. De formule zelf is:

$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}$

Hierin is de k de waarde waarvoor de kans op berekend voor wilt worden. Dit moet een positieve integer zijn dat groter is dan nul, aangezien deze in de faculteit van de formule moet. Lambda is hier de gemiddelde van de verdeling en de e hier is de constante van Euler.

```python
def poisson(lamb: float, x: int) -> float:
    """
    Deze functie implementeerd de functie voor de poisson verdeling
    :param lamb: De gemiddelde
    :param x: De waarde waarvan de kans berekend wilt worden.
    :return:
    Returnt de kans op gegeven waarde x in de vorm van een waarde van 0 tot 1.
    """
    return lamb**x * exp(-lamb)/factorial(x)
```

Deze wordt voor het aantal max_score gerund, zodat voor elke mogelijk aantal doelpunten de kans berekend wordt. Hiervan wordt de cumulatieve kansen van berekend, door de kans van het doelpunt plus de kans van hetzelfde aantal doelpunten min 1 te doen.

```python
def calculate_chance_each_score(lamb: float, max_score: int) -> dict:
    """
    Berekent voor elke mogelijke waarde de kans om die waarde te krijgen.
    :param lamb: De gegeven gemiddelde
    :param max_score: Het max aantal te behalen doelpunten
    :return:
    De dictionary waarin staat als key het aantal doelpunten met als value de cumulatieve kans op dat aantal doelpunten
    """
    chance_amount_score = {}
    for i in range(0, max_score):
        chance_amount_score[f"{i}"] = poisson(lamb, i)
    cum = {"-1": 0, "0": chance_amount_score["0"]}
    for i in range(1,20):
        cum[f"{i}"] = cum[f"{i-1}"] + chance_amount_score[f"{i}"]
    return cum
```

Deze functie berekent wat het aantal behaalde doelpunten is aan de hand de gegeven input, wat een willekeurig getal zal zijn tussen 0 en 1.

```python
def get_scores(input:float, cum: dict, max_score: int) -> int:
    """
    Deze functie berekent het aantal doelpunten gehaald aan de hand van input.
    :param input: Een willkeurig getal tussen 0 en 1
    :param cum: De dictionary waarin de cumulatieve kansen gegeven zijn per aantal doelpunt
    :param max_score: Maximaal aantal te behalen doelpunt
    :return:
    Het berekende aantal doelpunten van de team tijdens een wedstrijd.
    """
    for i in range(0, max_score):
        if cum[f"{i - 1}"] < input < cum[f"{i}"]:
            return i
        return max_score
```

Deze functie is een variant van de vorige win_status_check functie, maar geeft hier de score gebaseerd op het aantal doelpunten in de plaats van de willekeurige kans.

```python
def win_status_check_scores(score1: int, score2: int) -> tuple[int, int]:
    """
    Kijkt welke team gewonnen heeft gebaseerd op het aantal doelpunten dat de teams gescoord heeft.
    :param score1: Aantal behaalde doelpunten team 1
    :param score2: Aantal behaalde doelpunten team 2
    :return:
    """
    if score1 > score2:
        return 3, 0
    elif score1 == score2:
        return 1, 1
    else:
        return 0, 3
```

Deze functie wordt hier herschreven met de nieuwe functies die gebaseerd zijn op de doelsaldo's in de plaats van alleen een willekeurig getal. Dit zorgt ervoor dat de positions en ranking_chances functie niet ook herschreven hoeven te worden, maar wel dezelfde functionaliteit behoudt.

```python
def run_simulation() -> dict:
    """
    Deze functie runt de simulatie van een reeks aan wedstrijden een keer. De winstanden worden hier met kansen op aantal doelpunten berekend.
    :return:
    De dictionary waarin de key de naam van de team is en het aantal punten als value geeft.
    """
    points = {"Ajax": 0, "Feyenoord": 0, "PSV": 0, "FC Utrecht": 0, "Willem II": 0}
    max_score = 20
    for home, out_chances in percentage_table.items():
        for out, chances in out_chances.items():
            # De gemiddelde doelpunten die de thuis team zal scoren, is het gemiddelde van de gemiddelde doelpunten die de team scoret thuis en de gemiddelde doelpunten die de tegenstanders die uit spelen krijgt. Voor het gemiddelde van het aantal doelpunten die de team die uit speelt gaat scoren is het vice versa.
            lamb_home = (lambda_table[home]["HS"] + lambda_table[out]["AC"]) / 2
            lamb_out = (lambda_table[home]["HC"] + lambda_table[out]["AS"]) / 2
            score_home = get_scores(rng.next(), calculate_chance_each_score(lamb_home, max_score), max_score)
            score_out = get_scores(rng.next(), calculate_chance_each_score(lamb_out, max_score), max_score)
            result_home, result_out = win_status_check_scores(score_home, score_out)
            points[home] += result_home
            points[out] += result_out
    return points
```

Hieronder is de kanstabel te zien van alle teams voor elke plek, maar berekent aan de hand van doelpunten in de plaats van een willekeurig getal. Hier is te zien dat Ajax nog steeds bovenaan staat, maar heeft wel een lager percentage vergeleken met de vorige tabel. Verder zijn Feyenoord en PSV omgewisseld van posities op. Verder komen de resultaten over met de verwachting hier.

```python
ranking_chances(10000)
```
